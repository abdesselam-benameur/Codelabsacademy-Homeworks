# Fix the data

```python
with open("videosUS.csv", "r", encoding="UTF-8") as f:
    new_content = []
    cpt = 0
    for line in f:
        if line.startswith('\\n'):
            new_content[-1] = new_content[-1].strip() + line.replace('\\n', '')
        else:
            new_content.append(line.replace('\\n', ''))
    with open("videosUS-updated-version.csv", "w", encoding="UTF-8") as f:
        f.writelines(new_content)
    del new_content

df = pd.read_csv('videosUS-updated-version.csv')
df.to_csv("videosUS-updated-version.csv", sep = ';', index=False)
```

# 3. Spot the possible outliers in your dataset and get rid of them.

```python
for col in df:
    if df[col].dtype == np.float64:
        print(col)
```

```python
df_without_outliers = df.copy()
threshold = 3

cols = ["views", "likes", "dislikes", "comment_count"]

for col in cols:
    column = df_without_outliers[col] # get the column
    df_without_outliers[f"{col}_zscore"] = (column - column.mean()) / column.std() # z-score
    df_without_outliers = df_without_outliers[np.abs(df_without_outliers[f"{col}_zscore"] < threshold)] # remove outliers
```

# 6. Get how many tags you have and which are the most common tags.

```python
all_tags = set()
df_tags = df["tags"].dropna()
for tags in df_tags:
    tags = map(lambda t: (t).strip('" '), tags.split('|'))
    all_tags.update(tags)

if '' in all_tags:
    all_tags.remove('')
    
print(len(all_tags))
```
